## Cartoonifying Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the image.
image = cv2.imread("Ayushmaan.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
# We will be finding the edges of image.
grayImg = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
grayImg = cv2.medianBlur(grayImg, 5)
edges = cv2.adaptiveThreshold(grayImg, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 19,9)
```

```python
plt.imshow(edges, cmap = 'gray')
```

```python
# Cartoonifying the image.
color = cv2.bilateralFilter(image, 9, 250, 250)    # edge preserving smoothing filter.
cartoon = cv2.bitwise_and(color, color, mask = edges)
```

```python
# Displaying the image.
plt.imshow(cartoon)
```

```python
# Stylization - The stylization filter produces a output that looks like the image was painted using water color.
cartoon_image = cv2.stylization(image, sigma_s=150, sigma_r=0.25)  
plt.imshow(cartoon_image)  
```
